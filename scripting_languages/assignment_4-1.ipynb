# Data Visualzation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/city_temp.csv", index_col=[0])
df
```

```python
ax = df[["CityA_High", "CityA_Low"]].plot(
    kind="line",
    color=["red", "blue"],
    ylabel="Temperature",
    title="Temperature Over Time",
    marker="o",
)
ax.legend(["High", "Low"])
```

```python
df["CityA_Diff"] = df["CityA_High"] - df["CityA_Low"]

df[["CityA_Low", "CityA_Diff"]].plot(
    kind="bar",
    stacked=True,
    color=["blue", "red"],
    ylabel="Temperature",
)
```

```python
fig, ax = plt.subplots()

ax.scatter(df["CityA_High"], df["CityB_High"], color="red")
ax.scatter(df["CityA_Low"], df["CityB_Low"], color="green")

plt.show()
```
