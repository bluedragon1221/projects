# Data Visualzation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/city_temp.csv", index_col=[0])
df
```

```python
ax = df[["CityA_High", "CityA_Low"]].plot(
    kind="line",
    color=["red", "blue"],
    ylabel="Temperature",
    title="Monthly Temperatures",
    marker="o",
)
ax.legend(["High", "Low"])
```

```python
df["CityA.Diff"] = df["CityA.High"] - df["CityA.Low"]
df["CityB.Diff"] = df["CityB.High"] - df["CityB.Low"]

fig, ax = plt.subplots(1, 2, figsize=(12, 5))

def make_opts(num):
    ticks = range(0, 101, 10)
    colors = [["blue", "red"], ["darkblue", "darkred"]]

    return {
        "ax": ax[num],
        "kind": "bar",
        "stacked": True,
        "ylabel": "Temperature",
        "yticks": ticks,
        "color": colors[num]
    }

df[["CityA.Low", "CityA.Diff"]].plot(
    title="City A",
    **make_opts(0)
)

df[["CityB.Low", "CityB.Diff"]].plot(
    title="City B",
    **make_opts(1)
)

plt.show()
```

```python
fig, ax = plt.subplots()

ax.scatter(df["CityA_High"], df["CityB_High"], color="red")
ax.scatter(df["CityA_Low"], df["CityB_Low"], color="green")

plt.show()
```
