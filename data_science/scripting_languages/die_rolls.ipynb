```python
import numpy as np
import pandas as pd
import numpy.random as r
import matplotlib.pyplot as plt
```

```python
# This one will be uniform
flat_random = r.randint(1, 6, size=(1000))
df = pd.DataFrame(flat_random, columns=["nums1"])
df["sum"] = df.sum(axis=1)
df.groupby("sum")["nums1"].count().plot.bar()
```

```python
# This one will not
flat_random = r.randint(1, 20, size=(1000, 2))
df = pd.DataFrame(flat_random, columns=["nums1", "nums2"])
df["max"] = df.max(axis=1)
df["sum"] = df.sum(axis=1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

df.groupby("sum")["nums1"].count().plot.bar(ax=ax1)
ax1.title.set_text("Roll 2d20")

df.groupby("max")["nums1"].count().plot.bar(ax=ax2)
ax2.title.set_text("Take the higher of 2d20")
```
